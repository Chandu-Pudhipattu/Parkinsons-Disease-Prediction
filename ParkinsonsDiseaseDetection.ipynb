```python
# lets scale the data using minmaxscaler

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler((-1,1))
x = scaler.fit_transform(df_features)
```

```python
y = df_labels
```

```python
# train test split

from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
# lets implement Logistic Rrgression model

from sklearn.linear_model import LogisticRegression

logistic = LogisticRegression()
logistic.fit(x_train,y_train)
```

```python
# predictions and performance metrics

from sklearn.metrics import accuracy_score,confusion_matrix,classification_report

y_pred = logistic.predict(x_test)
print(accuracy_score(y_test,y_pred)*100)
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
# lets improve by implementing xgboost classifier algorithm

from xgboost import XGBClassifier

xgb = XGBClassifier(base_score=0.5,booster='gbtree',colsample_bylevel=1,
             colsample_bynode=1,colsample_bytree=1,gamma=0,learning_rate=0.1,max_delta_step=0,
             max_depth=3,min_child_weight=1,missing=np.nan,n_estimators=100,n_jobs=1,nthread=None,
             objective='binary:logistic',random_state=0,reg_alpha=0,reg_lambda=1,scale_pos_weight=1,
             seed=None,silent=None,subsample=1,verbosity=1)

xgb.fit(x_train,y_train)
```

```python
# predictions and performace metrics

y_pred = xgb.predict(x_test)
print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
# lets check for data overfitting using cross validation

from sklearn.model_selection import cross_val_score

cv_scores = cross_val_score(xgb, x_train, y_train, cv=5)  # 5-fold cross-validation
print(f"Cross-validation scores: {cv_scores}")
print(f"Mean CV score: {cv_scores.mean()}")
```

Cross-validation scores: [1. 0.92592593 0.88888889 1. 0.92592593] These are the accuracy scores for each of the 5 folds. In your case, the model performs very well across most of the folds, with one fold showing a slightly lower accuracy (0.88888889), but the others being 1.0 or 0.9259, indicating that the model is performing well across the different subsets of the data.

Mean CV score: 0.9481 This is the average performance of the model across all folds. A mean cross-validation score of 94.81% is a good indicator that your model generalizes well to unseen data and doesn't just memorize the training data (overfitting).

```python
## Hyper parameter Tuning with GridSearchCV

from sklearn.model_selection import GridSearchCV

# Define parameter grid
param_grid = {
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 5, 7],
    'n_estimators': [100, 200, 300]
}

grid_search = GridSearchCV(estimator=xgb, param_grid=param_grid, cv=5, scoring='accuracy')
grid_search.fit(x_train, y_train)

print(f"Best parameters: {grid_search.best_params_}")
print(f"Best cross-validation score: {grid_search.best_score_}")
```

the best cross-validation score is 95.56%. This indicates a very good model performance with the chosen hyperparameters, and it suggests that these values provide a better fit than the default settings.Even if we change max depth to 5 it wont change much.
